```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results_df = pd.read_csv("results-po-then-pa.csv")
```

```python
bio_pri_scr = results_df.query('model == "sen"').sort_values('epoch').private_score.reset_index(drop=True)
lan_pri_scr = results_df.query('model == "lan"').sort_values('epoch').private_score.reset_index(drop=True)
sen_pri_scr = results_df.query('model == "bio"').sort_values('epoch').private_score.reset_index(drop=True)
```

```python
rdm_pri_scr = [0.00180] * len(bio_pri_scr)
```

```python
plt.plot(bio_pri_scr, label = "bio (10 po then 10 pa)")
plt.plot(lan_pri_scr, label = "lan (10 po then 10 pa)")
plt.plot(sen_pri_scr, label = "sen (10 po then 10 pa)")
plt.plot(rdm_pri_scr, label = "rdm")
ax = plt.gca()
ax.set(xticks=range(0,21,2), ylim=[0.0,0.2], xlabel="epoch", ylabel="private score")
plt.legend()
plt.show()
```
